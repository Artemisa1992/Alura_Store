### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Sun the 'Preço' column to get the total revenue
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibir o faturamento de cada loja
print("Faturamento da Loja 1:", faturamento_loja1)
print("Faturamento da Loja 2:", faturamento_loja2)
print("Faturamento da Loja 3:", faturamento_loja3)
print("Faturamento da Loja 4:", faturamento_loja4)

# Calcular o faturamento total
faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4

# Exibir o faturamento total
print("Faturamento Total:", faturamento_total)

# Comparar o faturamento
faturamentos = {
    'Loja 1': faturamento_loja1,
    'Loja 2': faturamento_loja2,
    'Loja 3': faturamento_loja3,
    'Loja 4': faturamento_loja4,
}

loja_mais_lucrativa = max(faturamentos, key=faturamentos.get)
loja_menos_lucrativa = min(faturamentos, key=faturamentos.get)

print("Loja maior Faturamento:", loja_mais_lucrativa)
print("Loja menor Faturamento:", loja_menos_lucrativa)

import matplotlib.pyplot as plt

# Dados para o gráfico
lojas = list(faturamentos.keys())
valores_faturamento = list(faturamentos.values())

# Cores pastéis (exemplos, você pode ajustar)
cores_pasteis = ['#B3E2CD', '#FDCDAC', '#CBD5E8', '#F4CAE4']

import matplotlib.pyplot as plt

# Dados para o gráfico
lojas = list(faturamentos.keys())
valores_faturamento = list(faturamentos.values())

# Novas cores (códigos hexadecimais fornecidos)
novas_cores = ['#3A5F40', '#8B8589', '#0A2342', '#8DAACF']

# Criar o gráfico de barras com barras mais finas e as novas cores
plt.figure(figsize=(10, 6)) # Define o tamanho da figura

# Define a cor de fundo como cinza claro (mantendo a cor anterior se não for especificado o contrário)
# Se você quiser manter o fundo cinza mais escuro, use '#D3D3D3' ou outro cinza.
# Vamos manter o cinza claro original '#F0F0F0' para este exemplo, a menos que você prefira o mais escuro.
plt.gca().set_facecolor('#F0F0F0') # Cor de fundo

plt.bar(lojas, valores_faturamento, color=novas_cores, width=0.5) # Usa as novas cores
plt.title('Faturamento por Loja') # Define o título do gráfico
plt.xlabel('Lojas') # Define o rótulo do eixo X
plt.ylabel('Faturamento (R$)') # Define o rótulo do eixo Y
plt.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona grades no eixo Y
plt.show() # Exibe o gráfico

```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

# Supondo que os DataFrames loja, loja2, loja3, loja4 já foram carregados
# Concatena os DataFrames das quatro lojas em um único DataFrame
todas_lojas = pd.concat([loja, loja2, loja3, loja4], keys=['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])

# Agrupa por produto e soma o preço para encontrar o faturamento total por produto
faturamento_total_por_produto = todas_lojas.groupby('Produto')['Preço'].sum()

# Encontrar os 5 produtos mais vendidos no total
top_5_produtos = faturamento_total_por_produto.nlargest(5)

# Encontrar os 5 produtos menos vendidos no total (excluindo faturamento 0)
# Filtrando para remover produtos com faturamento zero antes de encontrar os menores
produtos_menos_vendidos = faturamento_total_por_produto[faturamento_total_por_produto > 0]
bottom_5_produtos = produtos_menos_vendidos.nsmallest(5)


print("Top 5 Produtos Mais Vendidos (por Faturamento Total):")
print(top_5_produtos)
print("\n" + "="*40 + "\n") # Separador

print("Top 5 Produtos Menos Vendidos (por Faturamento Total - excluindo faturamento zero):")
print(bottom_5_produtos)
```
